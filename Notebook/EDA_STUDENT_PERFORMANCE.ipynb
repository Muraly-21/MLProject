3.8 Adding columns for Total_Score and Average_Score

```python
df['total_score'] = df['math score'] + df['reading score'] + df['writing score']
df['average'] = df['total_score'] /3
df.head()
```

We can also remove the math_score,reading_score and writing_score if not needed.we kept here for additional information for the model

```python
reading_full = df[df['reading score'] == 100]['average'].count()
writing_full = df[df['writing score'] == 100]['average'].count()
math_full = df[df['math score'] == 100]['average'].count()

print(f"Number of students with  full marks in Reading : {reading_full}")
print(f'Number of students with full marks in Writing: {writing_full}')
print(f'Number of students with full marks in Reading: {math_full}')
```

Insights

From above values we get students have performed the worst in Maths

Best performance is in reading section
